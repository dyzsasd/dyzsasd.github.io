```python
import numpy as np
import tensorflow as tf
```

### 模拟线性函数观测值

```python
k = 2
n = 1000

w = np.array(range(0, k + 1))
x = np.hstack([np.ones((n, 1)), np.random.rand(n, k)])
y = np.dot(x, w) + np.random.rand(n) * 0.05
```

### 以上代码生成了二维平面

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(xs=x[:, 1], ys=x[:, 2], zs=y)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('y')
plt.show()
```

```python
import tensorflow as tf

X = tf.constant(x, name='X', dtype='float32')
Y = tf.constant(y, name='Y', dtype='float32')
XT = tf.transpose(X)

W = tf.reduce_sum(tf.transpose(tf.matmul(tf.matrix_inverse(tf.matmul(XT, X)), XT) * Y), axis=0)
```

运行程序

```python
with tf.Session() as sess:
    print( W.eval())
```
